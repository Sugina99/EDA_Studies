1. Mean/ Median/Mode replacement

```python
df = pd.read_csv('titanic.csv',usecols=['Age','Fare','Survived'])
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
median = df.Age.median()
median
```

```python
#  Function to fill median values

def impute_median(df,variable,median):
    df[variable+"_median"]=df[variable].fillna(median)
```

```python
impute_median(df,'Age',median)
```

```python
df
```

```python
df['Age_median'] = df['Age'].fillna(median)
```

### Advantages And Disadvantages of Mean/Median Imputation

#### Advantages
1. Easy to implement(Robust to outliers)
2. Faster way to obtain the complete dataset

#### Disadvantages
1. Change or Distortion in the original variance
2. Impacts Correlation

### 2. Random Sample Imputation

```python
df = pd.read_csv('titanic.csv',usecols=['Age','Fare','Survived'])
df.head(10)
```

```python
df.isna().sum()
```

```python
df.Age.dropna()
```

```python
df.Age.dropna().sample(df['Age'].isna().sum())
```

```python
df.Age.dropna().sample(df['Age'].isna().sum())
```

```python
s=df.Age.dropna().sample(df['Age'].isna().sum(),random_state=0)
```

```python
df['Age'].isnull().sum()
```

```python
df.Age.fillna(s)
```
