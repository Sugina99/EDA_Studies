```python
# dividing all features into diff category(num-continues,discrete  , object_type)
# numerical columns
num_cols=data.select_dtypes(np.number).columns
num_cols
```

```python
# object columns
object_cols=data.select_dtypes("object").columns
object_cols
```

```python
# continues and disceret columns
discrete_cols=[]
continuous_cols=[]
for feature in num_cols:
    if data[feature].nunique()>20:
        continuous_cols.append(feature)
    else:
        discrete_cols.append(feature)
print(f"continous columns are: {continuous_cols}")
print(f"discrete columns are: {discrete_cols}")
```
