```python
for column in df.columns:
    if df[column].isnull().sum()>0:
        print("The column",column,'has',df[column].isnull().sum(),"no. of missing values")
```

```python
# list comprehension
[column for column in df.columns if df[column].isna().sum()>0]
```

```python
df_country=pd.read_excel('ZomatoDataset/Country-Code.xlsx')
```

```python
df_country
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df['Country'].value_counts()
```

```python
names=final_df['Country'].value_counts().index
```

```python
names
```

```python
values=final_df['Country'].value_counts().values
```

```python
values
```

```python
plt.pie(values[:3],labels=names[:3],autopct='%1.3f%%')
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().sort_values(ascending=False)
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()  #form a data frame
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
from matplotlib import rcParams
rcParams['figure.figsize'] = 12,6

plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['White','Red','Orange','Yellow','Green','Green'])
```

```python
# count plot
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
